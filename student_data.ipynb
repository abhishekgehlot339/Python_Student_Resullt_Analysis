```python
df = df.drop("Unnamed: 0",axis=1)
print(df.head())
```

changed study hours column

```python
df["wklystudyhours"]=df["WklyStudyHours"].str.replace("10-may","10-05")
df.head()
```

gender distribution

```python
plt.figure(figsize=(4,4))
ax= sns.countplot(data=df,x ="Gender")
ax.bar_label(ax.containers[0])
plt.title("gender distribution")

plt.show()
```

from the above chart we have analysed that:
the number of females in data is more than the males data

```python
gb=df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
sns.heatmap(gb,annot=True)
plt.title("relationship between parent's education and student score")
plt.show()
```

from the above chart we have conclueded that the education of the parent 

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
sns.heatmap(gb1,annot=True)
plt.title("relationship between parent's maretial status and student's score")
plt.show()
```

from the above chart we have conclude that there is no negible imapct on the student score due their parant's maratial status
